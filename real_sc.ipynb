# Import the data set

Using pandas to import the data, here we're importing the Cho and Iyer data sets.

```python
import pandas as pd

cho_raw_df = pd.read_csv('./data/cho.txt', sep='\t', header=None)
iyer_raw_df = pd.read_csv('./data/iyer.txt', sep='\t', header=None)
```

# Data Preprocessing

Print the outliers percentage in the data set.

```python
iyer_outliers = iyer_raw_df[iyer_raw_df[1] == -1]

print(f'Percentage of outliers in Iyer dataset is {iyer_outliers.shape[0] * 100 / iyer_raw_df.shape[0]} %')
```

Remove outliers and extract labels

```python
iyer_cleaned_df = iyer_raw_df[iyer_raw_df[1] != -1]

cho_labels = cho_raw_df.iloc[:, 1]
cho_df = cho_raw_df.iloc[:, 2:]
print(f'Number of distinct labels in Cho dataset is {cho_labels.nunique()}')
iyer_labels = iyer_cleaned_df.iloc[:, 1]
iyer_df = iyer_cleaned_df.iloc[:, 1:]
print(f'Number of distinct labels in Iyer dataset is {iyer_labels.nunique()}')
```

# Spectral Clustering: Cho data set

Visualize the data set distribution range

```python
import matplotlib.pyplot as plt
cho_df.boxplot()
plt.title('Cho dataset')
plt.figure()
plt.show()
```

Perform Spectral Clustering on the Cho data set and visualize the results with the ground Truth labels.

```python
from sklearn.cluster import SpectralClustering
import numpy as np
import matplotlib.pyplot as plt

sc_iyer = SpectralClustering(n_clusters=5, affinity='nearest_neighbors', n_neighbors=10)
sc_iyer.fit(cho_df)
labels_cho = pd.DataFrame(sc_iyer.labels_, columns=['Cluster ID'])
result_cho = pd.concat((cho_df,labels_cho), axis=1)

plt.hist(result_cho['Cluster ID'])
plt.title('Spectral Clustering Clustering Distribution Results')
plt.show()

plt.hist(cho_labels)
plt.title('Ground Truth Distribution')
plt.show()
```

```python
mean = np.mean(result_cho['Cluster ID'], axis=0)
sse = np.sum((result_cho['Cluster ID'] - mean) ** 2)
print(f'SSE for Cho dataset is {sse}')
```

Visualize the confusion matrix for the Spectral Clustering results.

```python
from sklearn.metrics import confusion_matrix

k_labels = result_cho['Cluster ID']
k_labels_matched = np.empty_like(k_labels)

for k in np.unique(k_labels):
    match_nums = [np.sum((k_labels==k)*(cho_labels==t)) for t in np.unique(cho_labels)]
    k_labels_matched[k_labels==k] = np.unique(cho_labels)[np.argmax(match_nums)]

cm = confusion_matrix(cho_labels, k_labels_matched)

plt.imshow(cm,interpolation='none',cmap='Blues')
for (i, j), z in np.ndenumerate(cm):
    plt.text(j, i, z, ha='center', va='center')
plt.xlabel("Spectral Clustering label")
plt.ylabel("Truth label")
plt.show()
```

From the confusion matrix, we can see that the Spectral Clustering algorithm was able to cluster the majority of data points in clusters 0, 1, 3, 4. However, it was not able to cluster the data points in cluster 2. The best result was in the cluster 1, where the algorithm was able to cluster 119 out of 119 + 38 = 157 data points.

```python
print(f'Spectral Clustering accuracy for Cho dataset: {np.sum(k_labels_matched == cho_labels)*100 / len(cho_labels)} %')
```

Visualize the Spectral Clustering results on 2-D PCA.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(cho_df)
cho_df_2d = pca.transform(cho_df)

plt.scatter(cho_df_2d[:, 0], cho_df_2d[:, 1], c=k_labels_matched)
plt.title('Spectral Clustering Clustering Results')
plt.show()
plt.scatter(cho_df_2d[:, 0], cho_df_2d[:, 1], c=cho_labels)
plt.title('Ground Truth')
plt.show()
```

We see that the Spectral Clustering algorithm was able to cluster regions of the data set. However as the clusters are overlapping, the algorithm was not able to cluster all the data points that are in between the clusters.
